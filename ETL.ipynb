### [T] transformação dos dados

```python
Image("https://a.imagem.app/oXQxDl.png")
```

```python
contratos_mod = contratos.merge(empresas,
                               left_on = 'fk_empresa_contratada',
                               right_on = 'id_empresa',
                               how= 'left')
```

```python
contratos_mod
```

```python
contratos_mod.drop(columns= ["fk_empresa_contratada", "id_empresa"], inplace= True)
```

```python
contratos_mod.head(5)
```

```python
contratos_final = contratos_mod.merge(datas,
                                     left_on='inicio_vigencia',
                                     right_on= 'id_data',
                                     how= 'left')
```

```python
contratos_final.head(5)
```

```python
contratos_final.drop(columns= ['inicio_vigencia', 'id_data'], inplace= True)
```

```python
contratos_final
```

```python
contratos_final.rename(columns = {'data': 'data_inicio_vigencia'}, inplace= True)
```

```python
contratos_final.head(5)
```

```python
contratos_final = contratos_final.merge(datas,
                                 left_on= 'termino_vigencia',
                                 right_on= 'id_data',
                                 how= 'left')
```

```python
contratos_final.head(5)
```

```python
contratos_final.rename(columns = {'data': 'data_termino_vigencias'}, inplace= True)
```

```python
contratos_final.drop(columns= ['termino_vigencia', 'id_data'], inplace= True)
```

```python
contratos_final.head(5)
```

### tratamento dos dados

```python
contratos_final.count()
```

```python
contratos_final.dtypes
```

```python
contratos_final.data_inicio_vigencia = pd.to_datetime(contratos_final.data_inicio_vigencia,
                                                      format='%d/%m/%Y')
```

```python
contratos_final.data_termino_vigencias = pd.to_datetime(contratos_final.data_termino_vigencias,
                                                      format='%d/%m/%Y').dt.date
```

```python
for i in contratos_final.data_termino_vigencias:
    print(i)
    pd.to_datetime(i)
```

```python
contratos_final.data_termino_vigencias= contratos_final.data_termino_vigencias.str.replace("31/09/2017", "30/09/2017")
```

```python
contratos_final.data_termino_vigencias = pd.to_datetime(contratos_final.data_termino_vigencias,
                                                      format='%d/%m/%Y')
```

```python
contratos_final.dtypes
```

```python
contratos_final.head(5)
```

```python
contratos_final["tempo_contrato"] = (contratos_final.data_termino_vigencias - contratos_final.data_inicio_vigencia).dt.days
```

```python
contratos_final.head()
```

```python
contratos_final.tempo_contrato.value_counts()
```

```python
contratos_final = contratos_final[contratos_final.tempo_contrato > 0]
```

```python
contratos_final.tail()
```

```python
contratos_final.reset_index(drop= True, inplace= True)
```

```python
contratos_final.tail()
```

### [L] carregamento dos dados(Google Big Query)

```python
credentials = service_account.Credentials.from_service_account_file(filename="key.json", scopes = ["https://www.googleapis.com/auth/cloud-platform"])
```

```python
contratos_final.to_gbq(credentials= credentials, destination_table= "etl.etl_csv")
```
